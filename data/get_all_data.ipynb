```python
import pandas as pd
import vaex
from tqdm import tqdm
```

```python
!ls
```

```python
dfs = []

for i in tqdm(range(1,13)):
    ind = str(i) if i>=10 else ('0'+str(i))
    df = pd.read_parquet(f'./months/parquet/yellow_tripdata_2010-{ind}.parquet')
    dfs.append(df)
```

```python
merged_df = pd.concat(dfs)
```

```python
vaex_df = vaex.from_pandas(merged_df)
```

```python
def clean_rate_code(x):
    try:
        return int(x[0])
    except:
        return x
```

```python
vaex_df['rate_code'] = vaex_df['rate_code'].apply(clean_rate_code)
```

```python
vaex_df.export_hdf5('yellow_tripdata_2010_raw.hdf5', progress=True)
```
